```python
X = df[["rating", "rating_count","actual_price"]]
y = df["discounted_price"]


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

lr_model = LinearRegression()
lr_model.fit(X_train, y_train)

pred = lr_model.predict(X_test)

print('Linear Regression Intercept: ',lr_model.intercept_)
print('Linear Regression Coefficient: ',lr_model.coef_)
print('R2 Score: ', round(r2_score(y_test, pred),3))
```

```python
import plotly.graph_objects as go

fig1 = px.line(x=X_test["actual_price"], y=pred)
fig2 = px.scatter(x=X_train["actual_price"], y=y_train)

all_fig = go.Figure(data=fig1.data + fig2.data)
all_fig.update_layout(title='Actual Price by Discounted Price',
                     xaxis_title='Actual Price',
                     yaxis_title='Discounted Price')
all_fig.show()
```

```python
scaler = StandardScaler()

X = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3)

lr_model = LinearRegression()
lr_model.fit(X_train, y_train)

pred = lr_model.predict(X_test)
print("R2 Score: ", round(r2_score(y_test, pred),3))
```

```python
lasso = Lasso(alpha = 0.2)
lasso.fit(X_train, y_train)
lasso_pred = lasso.predict(X_test)

print('Lasso coefficients: ',lasso.coef_)
#print('R2score: ',r2_score(y_test, lasso_pred))
print('Lasso score: ',round(lasso.score(X_test, y_test), 3))
```

```python
ridge = Ridge(alpha = 0.1)
ridge.fit(X_train, y_train)
ridge_predict = ridge.predict(X_test)

print('Ridge coefficients: ',ridge.coef_)
#print('R2 score: ',r2_score(y_test, ridge_predict))
print('Ridge score: ',round(ridge.score(X_test, y_test),3))
```
